```python
import gradio as gr

# 1. Your original function (kept exactly the same for compatibility)
def chat_with_groq(message):
    chat_completion = client.chat.completions.create(
        messages=[
            {
                "role": "user",
                "content": message,
            }
        ],
        model="llama-3.3-70b-versatile",
    )
    return chat_completion.choices[0].message.content

# 2. Wrapper function to handle the chat history formatting for the modern UI
def respond(message, chat_history):
    # Get the response from your Groq function
    bot_message = chat_with_groq(message)
    # Append the user message and bot response to the history
    chat_history.append((message, bot_message))
    # Return an empty string (to clear the input box) and the updated history
    return "", chat_history

# 3. Custom CSS for a sleek, modern dashboard look
custom_css = """
.dashboard-header {
    text-align: center;
    padding: 20px;
    background: linear-gradient(135deg, #0f2027, #203a43, #2c5364);
    color: white;
    border-radius: 12px;
    margin-bottom: 20px;
    box-shadow: 0 4px 6px rgba(0,0,0,0.1);
}
.chat-container { border-radius: 12px !important; }
"""

# 4. Build the UI using gr.Blocks
with gr.Blocks(css=custom_css, theme=gr.themes.Soft(primary_hue="indigo", neutral_hue="slate")) as demo:

    # Modern Header
    gr.HTML(
        """
        <div class='dashboard-header'>
            <h1 style='margin: 0; font-size: 2em;'>⚡ My AI Enterprise Dashboard</h1>
            <p style='margin: 5px 0 0 0; opacity: 0.8;'>Powered by Llama 3.3 70B & DDS</p>
        </div>
        """
    )

    with gr.Row():
        # Left Sidebar for Controls/Info
        with gr.Column(scale=1, min_width=250):
            gr.Markdown("### ⚙️ Control Panel")
            gr.Markdown("Configure your session parameters.")

            # These are display elements to make it look professional
            gr.Dropdown(choices=["llama-3.3-70b-versatile"], value="llama-3.3-70b-versatile", label="Active Model", interactive=False)
            gr.Slider(minimum=0, maximum=1, value=0.7, label="Temperature (Display Only)", interactive=False)

            gr.Markdown("---")
            clear_btn = gr.Button("🗑️ Clear Conversation", variant="secondary")

        # Right Main Area for Chat
        with gr.Column(scale=3):
            # The Chatbot component for a professional message UI
            chatbot = gr.Chatbot(
                label="AI Assistant",
                height=500,
                elem_classes="chat-container",
                #bubble_full_width=False,
                avatar_images=(None, "https://groq.com/wp-content/uploads/2024/02/Groq-Logo-Black.png") # Adds the Groq logo to the bot
            )

            # Input Area
            with gr.Row():
                msg_input = gr.Textbox(
                    show_label=False,
                    placeholder="Type your message here and press Enter...",
                    scale=4,
                    container=False
                )
                submit_btn = gr.Button("Send ✨", variant="primary", scale=1)

    # 5. Wire up the functionality
    # Pressing Enter in the textbox
    msg_input.submit(respond, inputs=[msg_input, chatbot], outputs=[msg_input, chatbot])
    # Clicking the Send button
    submit_btn.click(respond, inputs=[msg_input, chatbot], outputs=[msg_input, chatbot])
    # Clicking the Clear button
    clear_btn.click(lambda: None, None, chatbot, queue=False)

# 6. Launch the dashboard
demo.launch(share=True)
```

```python
import os
import gradio as gr
from google.colab import userdata
from groq import Groq

# 1. Setup Groq Client
groq_api_key = userdata.get("GROQ_API_KEY")
client = Groq(api_key=groq_api_key)

# 2. History-Aware Chat Function
# Gradio passes the current message and the entire conversation history
def chat_with_groq(message, history):
    # Format history into the JSON structure Groq/Llama expects
    messages = [{"role": "system", "content": "You are a helpful, highly capable AI assistant."}]

    for user_msg, asst_msg in history:
        messages.append({"role": "user", "content": user_msg})
        messages.append({"role": "assistant", "content": asst_msg})

    # Append the newest user message
    messages.append({"role": "user", "content": message})

    # Call the API
    chat_completion = client.chat.completions.create(
        messages=messages,
        model="llama-3.3-70b-versatile",
    )
    return chat_completion.choices[0].message.content

# 3. Build Professional UI with Blocks
# Using a sleek slate theme optimized for technical dashboards
custom_theme = gr.themes.Soft(
    primary_hue="slate",
    neutral_hue="zinc",
    font=[gr.themes.GoogleFont('Inter'), 'ui-sans-serif', 'system-ui', 'sans-serif']
)

with gr.Blocks(theme=custom_theme, title="Groq AI Dashboard") as demo:
    # --- Header Section ---
    with gr.Row():
        gr.Markdown(
            """
            # ⚡ Groq Command Center
            ### High-Speed Inference Engine powered by Llama 3.3
            """
        )

    # --- Main Dashboard Layout ---
    with gr.Row():

        # Left Column: Information Card / Dashboard Panel
        with gr.Column(scale=1, variant="panel"):
            gr.Markdown("### 📊 System Dashboard")
            gr.Markdown("**Status:** 🟢 API Connected")
            gr.Markdown("**Provider:** Groq LPU")
            gr.Markdown("**Active Model:** Llama 3.3 70B")

            gr.Markdown("---")

            gr.Markdown("### 💡 Interface Features")
            gr.Markdown(
                """
                * **Context Aware:** The chatbot remembers previous messages in this session.
                * **Code Ready:** Hover over AI responses to reveal a **Copy** button.
                * **Advanced Controls:** Use the Undo, Retry, and Clear buttons integrated into the chat.
                * **Theme:** Toggle Light/Dark mode using the Gradio footer icon.
                """
            )

        # Right Column: Main Chat Interface (Panel variant for a clean card look)
        with gr.Column(scale=3, variant="panel"):
            # gr.ChatInterface automatically handles advanced buttons (Submit, Retry, Undo, Clear)
            gr.ChatInterface(
                fn=chat_with_groq,
                chatbot=gr.Chatbot(
                    height=600,
                    #show_copy_button=True,
                    layout="bubble", # Modern chat layout
                ),
                textbox=gr.Textbox(
                    placeholder="Type your prompt here...",
                    container=False,
                    scale=7
                ),
                examples=[
                    "Explain the importance of fast language models.",
                    "Write a Python script to deploy an AI agent.",
                    "Design a dark-mode UI layout structure."
                ]
                # Removed theme=custom_theme as gr.ChatInterface does not accept it directly
            )

# 4. Launch the App
demo.launch(debug=True)
```

```python
#Mohammad Arshad version
import gradio as gr
from google.colab import userdata
from groq import Groq

# =========================
# 1. Setup
# =========================

groq_api_key = userdata.get("GROQ_API_KEY")

if not groq_api_key:
    raise ValueError(
        "GROQ_API_KEY not found. Please add it in Google Colab Secrets and enable notebook access."
    )

client = Groq(api_key=groq_api_key)

MODEL_NAME = "llama-3.3-70b-versatile"


# =========================
# 2. Chat Function
# =========================

def chat_with_groq(message, history):
    """
    Gradio 6 passes history as OpenAI-style messages:
    [
        {"role": "user", "content": "..."},
        {"role": "assistant", "content": "..."}
    ]
    """

    messages = [
        {
            "role": "system",
            "content": "You are a helpful, clear, and practical AI assistant."
        }
    ]

    # Add previous conversation history
    if history:
        for item in history:
            role = item.get("role")
            content = item.get("content")

            # Only send clean text messages to Groq
            if role in ["user", "assistant"] and isinstance(content, str):
                messages.append({
                    "role": role,
                    "content": content
                })

    # Add latest user message
    messages.append({
        "role": "user",
        "content": message
    })

    response = client.chat.completions.create(
        model=MODEL_NAME,
        messages=messages,
        temperature=0.7,
        max_completion_tokens=1024,
    )

    return response.choices[0].message.content


# =========================
# 3. Gradio App
# =========================

demo = gr.ChatInterface(
    fn=chat_with_groq,
    title="⚡ Groq AI Assistant",
    description="A simple Gradio chatbot powered by Groq and Llama 3.3.",
    examples=[
        "Explain Groq in simple terms.",
        "Give me 5 use cases of fast LLM inference.",
        "Write a Python function to clean a dataset.",
        "Explain RAG in beginner-friendly language."
    ],
    fill_height=True,
    fill_width=False,
)


# =========================
# 4. Launch
# =========================

demo.launch(
    theme=gr.themes.Soft(),
    debug=True
)
```
