```python

import os
from openai import OpenAI
import gradio as gr

# Load API key from Colab Secrets
from google.colab import userdata
os.environ["OPENAI_API_KEY"] = userdata.get("openai")

# Create OpenAI client
client = OpenAI()

# Function to ask the app
def ask_app(question):
    try:
        response = client.responses.create(

            prompt={
                #change this id
                "id": "pmpt_69de5e18b8a08196abdbbd45296bcef105bb256b91efc3ca",
                "version": "1"
            },
            input=question,
            reasoning={
                "summary": "auto"
            },
            tools=[
                {
                    "type": "file_search",
                    "vector_store_ids": [

                #change this id
                        "vs_69de5cdc30248191a00f60980b7c8433"
                    ],
                    "max_num_results": 5
                }
            ],
            store=True
        )
        return response.output_text
    except Exception as e:
        return f"Error: {str(e)}"

# Gradio interface
demo = gr.Interface(
    fn=ask_app,
    inputs=gr.Textbox(
        lines=3,
        placeholder="Type your question here...",
        label="Your Question"
    ),
    outputs=gr.Textbox(
        lines=10,
        label="Answer"
    ),
    title="Anusha Vaibhav Bot",
    description="Ask a question and get answers from your saved prompt + vector store."
)

demo.launch(share=True)
```
